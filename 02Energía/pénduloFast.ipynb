## Quiero las energías cinética y potencial

pénduloHorizontal.png

Importamos lo que necesitamos del módulo `SymPy`.


```python
import sympy as sm # importamos funciones de cálculo simbólico
from sympy.physics import mechanics as me # de sympy utilizaremos funciones de mecánica
me.init_vprinting() # notación con punto para la velocidad y punto punto para la aceleración
```

Definimos los parámetros físicos del sistema.

```python
g, m, l = sm.symbols(r'g, m, \ell', positive=True)
```

Coordenada generaliza.

```python
fi = me.dynamicsymbols(r'\varphi')
```

Un marco de referencia.

```python
e = me.ReferenceFrame('e')
```

Vector posición.

```python
m_r = l* (sm.cos(fi)* (-e.y) + sm.sin(fi)*e.x)
```

Aceleración gravitatoria.

```python
g_vector = g * (-e.y)
```

Y estaríamos listos para calcular las energías cinética y potencial del péndulo ideal.
