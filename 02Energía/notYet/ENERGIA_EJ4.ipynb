```python
import sympy as sp
from sympy.physics import mechanics as me

me.init_vprinting()

# Variables y Símbolos
t = sp.symbols('t')
theta = me.dynamicsymbols('theta')
m1, m2, g, a, Omega = sp.symbols('m1 m2 g a Omega')

# Marcos de Referencia
N = me.ReferenceFrame('N') # Inercial
A = me.ReferenceFrame('A') # Rota con el sistema
A.orient_axis(N, Omega * t, N.y) # Gira en torno al eje vertical Y
A.set_ang_vel(N, Omega * N.y)

# Puntos y Cinemática
O = me.Point('O') # Punto fijo A (tope superior)
O.set_vel(N, 0)

# Masa m1 (una de las laterales)
# Posición: baja 'a*cos(theta)' y se aleja 'a*sin(theta)'
# Usamos el marco A porque m1 rota CON el sistema
r_m1 = a * sp.sin(theta) * A.x - a * sp.cos(theta) * A.y
M1 = O.locatenew('M1', r_m1)
M1.set_vel(N, r_m1.dt(N))

# Masa m2 (la inferior sobre el eje)
# Su posición es el doble de la caída vertical de m1
# Importante: m2 está en el eje, su velocidad es solo vertical en N
r_m2 = -2 * a * sp.cos(theta) * N.y
M2 = O.locatenew('M2', r_m2)
M2.set_vel(N, r_m2.dt(N))

# Energías
# Energía Cinética Total: 2 veces m1 + 1 vez m2
v1 = M1.vel(N)
v2 = M2.vel(N)

T_m1_individual = (sp.Rational(1, 2) * m1 * v1.dot(v1))
T_m2 = (sp.Rational(1, 2) * m2 * v2.dot(v2))

T_total = 2 * T_m1_individual + T_m2
T_final = sp.simplify(T_total)

# Energía Potencial (Tomando y=0 en el punto A superior)
# Definimos las ALTURAS (y) de forma explícita respecto al origen A
# Como el eje y apunta hacia arriba y las masas están abajo, son negativas.
h_m1 = -a * sp.cos(theta)
h_m2 = -2 * a * sp.cos(theta)

# Las masas están por debajo de O (referencia), por eso el producto será negativo (ESTO EXPLICA QUE LA ENERGÍA POTENCIAL GRAVITATORIA TIENE UN SIGNO MENOS DEBIDO A QUE LAS ALTURAS SE UBICAN EN LA PARTE NEGATIVA DEL EJE Y )
V_total = 2 * (m1 * g * h_m1) + (m2 * g * h_m2)
V_final = sp.simplify(V_total)

# Energía Cinética de una de las masas m1

# Incluye la rotación alrededor del eje y la oscilación del brazo
T_m1 = sp.simplify(sp.Rational(1, 2) * m1 * v1.dot(v1))

# Energía Cinética de la masa m2
# Solo tiene movimiento vertical (traslación pura en el eje del regulador)
T_m2_final = sp.simplify(sp.Rational(1, 2) * m2 * v2.dot(v2))

print("Energía Cinética de una masa m1:")
display(T_m1)

print("\nEnergía Cinética de la masa m2:")
display(T_m2_final)

print()

T_tras = sp.symbols(r'T_{\text{traslación}}')
V_grav = sp.symbols(r'V_{\text{gravitatoria}}')

display(sp.Eq(T_tras, T_final))
display(sp.Eq(V_grav, sp.factor(V_final)))
```
