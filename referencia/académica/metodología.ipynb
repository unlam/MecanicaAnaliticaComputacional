## Código para resolver problemas de mecánica

- Se escribirá código usando el lenguaje de programación general [Python](https://www.python.org/)
- Escribir código $\neq$ programar
    - **Programar**: Implica analizar y modelar matemáticamente un problema (lógica), idear y comparar alternativas, cuál es la mejor considerando contexto de uso, restricciones, arquitectura del equipo informático y capacidades del humano que le opere.
        - Requiere **bastante** estudio y esfuerzo.
        - Cualquier estudiante de ingeniería **puede** aprender a programar.
    - **Escribir código**: Permite obtener soluciones informáticas valíendose de un lenguaje de programación.
        - Requiere **poco** estudio y esfuerzo
        - Todo estudiante de ingeniería **DEBE** aprender a escribir código.
        - Grandes modelos de lenguaje facilitan la tarea: [ChatGPT](https://chat.openai.com/), [DeepSeek](https://chat.deepseek.com/), [Mistral](https://chat.mistral.ai/), [Qwen](https://chat.qwen.ai/), [Claude](https://claude.ai), [Copilot](https://github.com/features/copilot), etc.
- Se emplea un mínimo de funciones de bibliotecas:
    - [SymPy](https://www.sympy.org/es/index.html) para álgebra y análisis simbólico
    - [NumPy](https://numpy.org/) y [SciPy](https://scipy.org/) para cálculo numérico
    - [matplotlib](https://matplotlib.org/) para generar representaciones visuales de los resultados

- Los _cuadernos_ de JupyterLab se despliegan en un navegado, incluso en un teléfono.
- Están segmentados en celdas de:
    - **anotación**: textos, figuras, multimedia y simbología en notación matemática estándar,
    - **código**: cálculo simbólico, representación gráfica y simulación numérica.
- Esto hace de Los cuadernos un soporte ideal para entregar a los alumnos:
    - El material teórico.
    - Ejercicios resueltos.
- Son **reutilizables**, esto es, a los alumnos les basta copiarles y modificarles para:
    - Resolver cada ejercicio de la guía de problemas.
    - Hacer una devolución de los mismos al docente para la evalución.

## [Aula invertida](https://es.wikipedia.org/wiki/Aula_invertida)

diagramaTiempo.png

- Teoría
    - Se lee, e intenta aplicar en la resolución de ejercicios, **antes** del encuentro semanal.
    - A través de un foro en línea se puede hacer consultas en todo momento.
    - Cualquier duda remanente debe aclararse **durante** el encuentro.
- Ejercicios
    - Se inician **antes** del encuentro.
    - Debe finalizarse **durante** el mismo, con ayuda del docente.
    - La corrección se realiza **tras** el encuentro.

### Organización del tiempo

- Esta modalidad requiere del alumno **mucha regularidad**.
- Si espera al encuentro sincrónico para:
  - leer teoría $\implies$ **no alcanza el tiempo**
  - empezar los ejercicios $\implies$ **no alcanza el tiempo**
  - ponerse _al día_ con lo adeudado de la semana anterior $\implies$ **no alcanza el tiempo**
- Cada semana:
  - **Nueva guía** con ejercicios marcados para su entrega 
  - **Entrega obligatoria** vía Teams 10' antes del cierre de cada encuentro sincrónico
- Por supuesto que puede fallarse en uno o más de los ejercicios que se entregan.
  - Es posible que haya faltado comprensión sobre algún aspecto de teoría $\implies$ consulta y/o repaso personal de teoría
  - Pero tal falla debe corregirse en una **segunda entrega**.

### Asistencia
- En encuentros sincrónicos se toma lista.
- Se requiere un 75% de asistencia para aprobar.
- Sí faltan por cuestiones médicas, un certificado permite descontarle.

### Entrega de ejercicios
- Cada unidad tiene su guía de ejercicios.
    - Deben entregarse la resoluciones de los que aparecerán en _Teams_ bajo _Tareas_.
- El día que se presenta la unidad podrán comenzar a mirarlos y resolverlos.
    - En el siguiente encuentro podrán realizar consultas adicionales
    - Los ejercicios se entregan antes de que comience el siguiente.
- Los encuentros sincrónicos son para consultas y finalizar ejercicios.
    - También pueden consultar en Teams durante la semana.

#### Esquema de entregas
> La guía de ejercicios N se publica en la semana N-1 $\rightarrow$ en N se entregan

- Pasado el día de entrega $\implies$ penalización sobre la nota
    - Por ello es importante que intenten avanzar en ellos, no es necesario que estén perfectos, una vez los envíen se los corregiremos y les pediremos las modificaciones necesarias para que puedan avanzar hasta que lo terminen satisfactoriamente.
    - Cuando termine la cursada entonces deberían haber logrado hacer **TODOS** los ejercicios asignados como _Tarea_.
