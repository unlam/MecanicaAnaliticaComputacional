## Fuerza y momento

> Referencia:
>- Sección 1.4  
John Robert Taylor, Classical Mechanics (University Science Books, 2005).

Como nuestra experiencia cotidiana nos enseña, podemos imprimir una aceleración a un objeto aplicándole una **fuerza**, $\vec{F}$.
Y también la experiencia nos enseña que está deberá ser mayor si lo es la masa del objeto, $m$, si queremos que la aceleración sea la misma.
Así que $m$ no es más que un factor de proporcionalidad entre la fuerza y la aceleración.

Con esta idea, la 2.a ley de Newton que puede resumirse en
$$
\dfrac{\vec{F}}{m} = \ddot{\vec{r}}
$$,

Si aceptamos que algo intangible como la aceleración de un campo gravitatorio, $\vec{g}$, imprime una variación en el tiempo de $\dot{\vec{r}}$ a un cuerpo bajo su acción, puede decirse algo similar de la acción de que una fuerza externa, $\vec{F}$.

En lo cotidiano, donde las masa nunca cambia, se puede decir que algo tangible como la $\vec{F}$ le imprime una variación en el tiempo
$$
\boxed {
\vec{F} =
m \dfrac{\partial}{\partial t} (\dot{\vec{r}}) = 
\dfrac{\partial}{\partial t} (m \dot{\vec{r}}) = \dot{\vec{p}} 
} ,
$$
a una cantidad que recibe el nombre de **momento**, $\vec{p} = m \dot{\vec{r}}$.

## Momento angular

Para un objeto puntual con momento $\vec{p}$, desde un punto de referencia $O$, siempre podrá definirse un sistema de coordenadas $\hat{e}_1,\hat{e}_2,\hat{e}_3$ centrado en $O$, tal que se pueda expresar el vector de posición $\vec{r}$ perpendicular a $\vec{p}$,   
$$
\begin{aligned}
\vec{r} = \rho \, \hat{e}_1 ,\\ 
\vec{p} =  m v \, \hat{e}_2 ,
\end{aligned}
$$
de forma de obtener el tercer vector, el del momento angular en la dirección $\hat{e}_3$: 
$$
\vec{L}_O = \rho \, \hat{e}_1  \times m v \, \hat{e}_2 = \rho m v \, (\hat{e}_1  \times \hat{e}_2) = \rho m v \, \hat{e}_3
.
$$
El símbolo $\times$ denota la operación de **producto vectorial** entre vectores que produce un tercero perpendicular a los anteriores.

Por ejemplo en el caso de un movimiento que implique una rotación circular, podemos esquematizar la relación entre vectores como en la figura siguiente.

imagen.png

Formalmente, se define al momento angular de una masa puntual se define como
$$
\vec{L}_O(t) = 
\vec{r} \times (m \dot{\vec{r}} ) =
\vec{r}(t) \times \vec{p}(t) 
.
$$ 

Es importante recordar que el subíndice $_O$ indica que depende del punto de referencia (centro de coordenadas) que se seleccionó para definirlo, aunque en general se lo resume en
$$
\boxed {
\vec{L} = \vec{r} \times \vec{p} 
} .
$$

#### Orientación del producto vectorial

La convención y regla mnemotécnica de la **regla de la mano derecha** es útil para recordar la dirección en el espacio tridimensional que asumirá el producto vectorial $\vec{a} \times \vec{b}$ entre dos vectores $\vec{a}$ y $\vec{b}$.

[Right hand rule cross product.svg](https://en.wikipedia.org/wiki/File:Right_hand_rule_cross_product.svg)

### Variación del momento angular

> Referencia:
>- Sección 2.5  
Stephen T. Thornton and Jerry B. Marion, Classical Dynamics of Particles and Systems, 5th ed (Thomson Brooks/Cole, 2004).

Ver la variación en el tiempo de $\vec{L}$, que si bien es un producto vectorial, es un producto a fin de cuentas requiere de usar la regla de la cadena
$$
	\dot{\vec{L}} = \dot{\vec{r}} \times \vec{p} + \vec{r} \times \dot{\vec{p}}
.
$$

Pero en el primer término sucecede algo interesante, ya que el momento lineal $\vec{p}$ no es más que multiplicar una constante por la velocidad, esto hace que sean vectores paralelos.
Es decir
$$ \vec{p}= m \dot{\vec{r}} \implies \vec{p} || \dot{\vec{r}}$$

Y el producto vectorial de dos vectores paralelos es cero
$$\vec{p} || \dot{\vec{r}} \implies \dot{\vec{r}} \times \vec{p} = 0 $$

En consecuencia en nuestro cálculo de la variación en el tiempo del momento angular, se cancela el primer término
$$
\dot{\vec{L}} = \not\dot{\vec{r}} \not\times \not\vec{p} + \vec{r} \times \dot{\vec{p}}
,
$$
<!---
$$
	\dot{\vec{L}} = \cancel{ \dot{\vec{r}} \times \vec{p}} + \vec{r} \times \dot{\vec{p}}
	,
%$$
--->
con lo que resulta que
$$
\boxed{
	\dot{\vec{L}} = \vec{r} \times \dot{\vec{p}}
}
.
$$

## Torque

Así como el causante de una variación del momento, $\dot{\vec{p}}$, recibía el nombre de fuerza, $\vec{F}$, la variación en el tiempo del momento angular, $\dot{\vec{L}}$, también recibe un nombre particular, torque, 
$$
  \vec{\tau}_O = \dot{\vec{L}}= \vec{r} \times \dot{p} = \vec{r} \times \vec{F}
  ,
$$
que nuevamente está referido a un punto ("$O$") y resulta entonces perpendicular a la fuerza $\vec{F}$ y $\vec{r}$, el vector que va desde la posición "$O$" hasta el punto donde de aplicación de la fuerza.

$$
\boxed{
  \vec{\tau} = \vec{r} \times \vec{F}
}
$$

imagen.png

El torque queda definido como el producto vectorial entre el vector posición y la fuerza aplicada.
Por las propiedades del producto vectorial, $\vec{\tau}$, es perpendicular al plano formado por estos dos vectores.

- $\vec{r}$ es nu vector posición que va desde el eje de rotación hasta el punto de aplicación de la fuerza y recibe el nombre de **brazo de palanca**.
- $\vec{F}$ es la fuerza aplicada.
 
Si para $\vec{r}$ la unidad de medida es el metro (m) en el Sistema Internacional de Unidades (SI), y son los newtons (N) para $\vec{F}$, el torque $\vec{\tau}$ se mide en newton-metros (N·m) en el SI.

Hay que recordar que el brazo de palanca llega hasta la posición donde se aplica la fuerza. Por esto era que si queremos abrir una puerta, por ejemplo, tenemos que aplicar más fuerza si la abrimos tirando cerca del eje de giro que lejos de él. Para lograr el mismo torque tenemos que aumentar la fuerza porque el brazo de palanca cambió. 

### Cálculo del vector torque

```python
# biblioteca de cálculo simbólico
import sympy as sm
import sympy.physics.mechanics as me
me.init_vprinting() # notación con puntos para derivadas temporales
```

El establecer un sistema de coordenadas, por ejemplo, cartesiano, define que signos tendrán las magnitudes vectoriales, como el torque, en sus componentes vectoriales.

```python
# Sistema cartesiano
e = me.ReferenceFrame('e') # marco referencial e en coordenadas cartesianas
```

Tomando la ilustración de arriba, podemos imaginar unos ejes tales que $\vec{r} = r \hat{e}_y$ y $\vec{F} = F (- \hat{e}_x)$.

Esto es equivalente a decir

$\vec{F} = F (- \hat{e}_x) = -F \hat{e_x} + 0 \hat{e}_y + 0 \hat{e}_z $

$\vec{r} = r (- \hat{e}_y) = 0 \hat{e_x} + r \hat{e}_y + 0 \hat{e}_z $

```python
F, r = sm.symbols('F r') # magnitudes escalares
fuerza = F* (-e.x)
punto_aplicación = r* e.y

torque = punto_aplicación.cross(fuerza) # momento de fuerza
torque
```

Lo que resulta en un $\vec{\tau} = F r \hat{e}_z$ como muestra la figura, positivo en $\hat{e}_z$.

La forma de calcular el torque es la misma en cualquier sistema de coordenadas.:
$$
\vec{\tau} =
\vec{r} \times \vec{F} = 
\begin{vmatrix}
\hat{e}_i & \hat{e}_j & \hat{e}_k \\
r_i & r_j & r_k \\
F_i & F_j & F_k
\end{vmatrix}
= \hat{e}_i (r_j F_k - r_k F_j) + \hat{e}_j (r_k F_i - r_i F_k) + \hat{e}_k (r_i F_j - r_j F_i)
$$


La cuenta que realizamos recién es 

$$
\vec{\tau} =
\vec{r} \times \vec{F} = 
\begin{vmatrix}
\hat{e}_x & \hat{e}_y & \hat{e}_z \\
0 & r & 0 \\
-F & 0 & 0
\end{vmatrix}
= \hat{e}_x (0 - 0) + \hat{e}_y (0 - 0) + \hat{e}_z (0 - (-Fr)) = Fr \hat{e}_z
$$

#### Invirtiendo uno de los vectores

Si se invierte alguno de los vectores, obviamente lo hará la dirección de $\vec{\tau}$.

```python
fuerza = F* e.x
punto_aplicación = r* e.y
torque = punto_aplicación.cross(fuerza) # momento de fuerza
torque 
```

La cuenta que realizamos recién es 

$$
\vec{\tau} =
\vec{r} \times \vec{F} = 
\begin{vmatrix}
\hat{e}_x & \hat{e}_y & \hat{e}_z \\
0 & r & 0 \\
F & 0 & 0
\end{vmatrix}
= \hat{e}_x (0 - 0) + \hat{e}_y (0 - 0) + \hat{e}_z (0 - (F r)) = - F r \hat{e}_z
$$

#### Y esto es así para cualquier sistema de coordenadas

En un sistema de coordenadas cilíndricas, si aplico fuera en el sentido positivo de $\hat{e}_\varphi$, es decir se aplica en un sentido anti-horario para una rotación en torno al eje $\hat{e}_z$, el torque que provoca es positivo en dicho eje.

```python
cyl = me.ReferenceFrame('c', indices=(r"\rho", r"\varphi", "z"))
```

```python
ro = sm.symbols('rho') # coordenada radial
fuerza = F* cyl.y # fuerza en coordenadas cilíndricas
punto_aplicación = ro* cyl.x

punto_aplicación, fuerza
```

```python
torque = punto_aplicación.cross(fuerza) # momento de fuerza
torque
```

Es decir 
$$
\vec{\tau} =
\vec{r} \times \vec{F} = 
\begin{vmatrix}
\hat{e}_\rho & \hat{e}_\varphi & \hat{e}_z \\
\rho & 0 & 0 \\
0 & F & 0
\end{vmatrix}
= \hat{e}_\rho (0 - 0) + \hat{e}_\varphi (0 - 0) + \hat{e}_z (\rho F - 0) =  F \rho \hat{e}_z
$$

Y lo opuesto se obtiene si la fuerza favorece una rotación en sentido horario, es decir, si se aplica en el sentido negativo de $\hat{e}_\varphi$.  

```python
fuerza = - F* cyl.y # fuerza en coordenadas cilíndricas
punto_aplicación = ro* cyl.x
punto_aplicación, fuerza
```

```python
torque = punto_aplicación.cross(fuerza) # momento de fuerza
torque
```

Es decir 
$$
\vec{\tau} =
\vec{r} \times \vec{F} = 
\begin{vmatrix}
\hat{e}_\rho & \hat{e}_\varphi & \hat{e}_z \\
\rho & 0 & 0 \\
0 & -F & 0
\end{vmatrix}
= \hat{e}_\rho (0 - 0) + \hat{e}_\varphi (0 - 0) + \hat{e}_z (\rho (-F)  - 0) =  - F \rho \hat{e}_z
$$

## Momento angular y torque de cuerpos rígidos

Luego dejaremos de analizar objetos como partículas puntuales sino que los trataremos como cuerpos extensos rígidos.
En las materias básicas de física se establece que:
$$
\vec{L} = I \vec{\omega}.
$$

¿Cómo cambia esto con el tiempo?
$$
\dot{\vec{L}} = \dot{I} \vec{\omega} + I \dot{\vec{\omega}} 
.
$$

Pero si el cuerpo es rígido, es decir no puede deformarse, el momento angular no puede cambiar con el tiempo, por lo que:
$$
\dot{\vec{L}} = \not\dot{I} \not\vec{\omega} + I \dot{\vec{\omega}}
,
$$
y por tanto el torque es:
$$
\vec{\tau} =\dot{\vec{L}} = I \dot{\vec{\omega}}
.
$$

### Conservación del momento angular

Así como decíamos que si la suma de fuerzas era nula entonces el momento se conservaba, encontramos un teorema equivalente para el momento angular a partir de los torques.

De la misma forma que para la suma de fuerzas, en el caso de la suma de torques vale

$$
\vec{\tau}_{\text{total}} = \sum_i \vec{\tau}_i = \dot{\vec{L}}
.
$$

Si $\vec{\tau}_\text{total} = 0 \implies \vec{L} = \text{constante}$.

## Un ejemplo

Imaginemos una barrera motorizada para impedir el paso de automotores.
La barrera es un cilindro de longitud $\ell$ y radio despreciable frente a esta magnitud.
Esta puede elevarse un ángulo $\theta$ arbitrario respecto a la horizontal gracias al torque que aplica un motor en su extremo fijo.

image.png

### Torque para compensar el peso

Como puede verse en la ilustración, la fuerza peso actúa en el centro de masa de la barrera.
Si el motor no ejerciera un torque, esta no podría mantener un $\theta$ constante.

Este $\theta \equiv \mathrm{constante}$ implica una nula velocidad angular, $\dot{\theta} = \omega = 0$.
Es decir, un momento angular también nulo, pues $\vec{L} = I \vec{\omega} = 0$.

Y si $\vec{L} \equiv \mathrm{constante}$, es decir que no varía en el tiempo, $\dot{\vec{L}} = I \dot{\omega} = 0$, por lo visto anteriormente, entonces la suma de torques debe ser nula, $\vec{\tau}_{\text{total}} = \sum_i \vec{\tau}_i = 0$.

Puesto que siempre que la barrera no esté vertical, la fuerza peso genera un torque, $\vec{\tau}_\text{peso} \neq 0$, entonces debe existir un torque del motor que compense este torque de la fuerza peso.

```python
m, l = sm.symbols('m, ell') # parámetros del cilindro
zeta = sm.symbols('theta') # coordenada angular
g = sm.symbols('g') # aceleración de la gravedad
vector_gravedad = g* -(e.y) # vector de gravedad en coordenadas cilíndricas
centro_masa = (l/2)* (sm.cos(zeta)* e.x + sm.cos(zeta)* e.y) # centro de masa en coordenadas cilíndricas
torque_peso = centro_masa.cross(m * vector_gravedad) # momento de fuerza debido a la gravedad
torque_peso
```

Todo cuanto debe hacer el motor es proveer:

```python
torque_motor_peso = - torque_peso # momento de fuerza del motor
torque_motor_peso
```

### Torque dinámico

¿Qué pasa cuando se quiere cambiar $\theta$?
Se parte desde algún $\theta$ en que la barrera está estática y se quiere cambiar a otro $\theta'$.
Obviamente, para esto se requiere alcanzar algún $\dot{\theta} \neq 0$, y como se parte y arriba a $\dot{\theta} = 0$, evidéntemente hay que acelerar la barrera, es decir hay momentos donde $\ddot{\theta} = \dot{\omega} \neq 0$.

Se $\dot{\omega} \neq 0 \implies \vec{L} \neq 0 \implies \sum_i \vec{\tau}_i = 0$, habrá un torque neto distinto de cero, $\vec{\tau}_{\text{total}} \neq 0$.

```python
barra_I_cm  = sm.Rational(1,12) * m * l**2
steiner_extremo_cm = m* (l* sm.Rational(1/2))**2 
barra_I_extremo = barra_I_cm + steiner_extremo_cm # teorema de Steiner
barra_I_extremo.simplify()
```

```python
omega_p = sm.symbols('\dot{\omega}') # velocidad angular de la barrera
torque_motor_dinámico = barra_I_extremo * omega_p # momento de fuerza dinámico
torque_motor_dinámico
```

Recordemos que esto no es nulo solo cuando hay una aceleración angular, $\dot{\omega} \neq 0$.
Si se establece una velocidad angular constante, $\omega \equiv \mathrm{constante} \implies \dot{\omega} = 0$, este torque dinámico es nulo.

### Torque total para la barrera

El torque total que el motor debe suministrar dependerá del ángulo de la barrera $\theta$ y de si está cambiando tal ángulo o no.
En este último caso **solo hay un término adicional si hay aceleración angular**, esto es $\alpha = \dot{\omega} = \ddot{\theta} \neq 0$.

```python
sm.Eq(
    sm.symbols(r'\vec{\tau}_{motor}'), 
    torque_motor_peso.dot(e.z) + torque_motor_dinámico,
    evaluate=False   
)
```

Si la barrera está cambiando $\theta$, pero a **velocidad angular constante**, $\omega = \mathrm{constante} \implies \dot{\omega} = 0$, **no hay torque dinámico** adicional al requerido para equilibrar el generado por el peso de la barrera.
