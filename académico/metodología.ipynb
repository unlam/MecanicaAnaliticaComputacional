## Código para resolver todos los problemas

- Escribiremos código del lenguaje de programación general [Python](https://www.python.org/)
- Escribir código $\neq$ programar
    - **Programar**: Saber cómo analizar y modelar matemáticamente un problema (lógica), idear y comparar alternativas y saber calcular cuál es la mejor considerando contexto de uso, restricciones, arquitectura del sistema informático, así como cuestiones de interfaz con el operador humano, etc.
        - Requiere **bastante** estudio y esfuerzo.
        - Cualquier estudiante de ingeniería mecánica **puede** aprender a programar y sacar provecho profesional de esto.
    - **Escribir código**: Plasmar una solución en un lenguaje de programación particular.
        - Requiere **poco** estudio y esfuerzo, más contando con la ayuda de grandes modelos de lenguaje como [ChatGPT](https://chat.openai.com/), [DeepSeek](https://chat.deepseek.com/) o [Copilot](https://github.com/features/copilot).
        - Todo estudiante de ingeniería mecánica **DEBE** escribir código para aprovecharlo en todas las asignaturas y también luego en su carrera profesional.
- Usaremos un mínimo de funciones de bibliotecas
    - [SymPy](https://www.sympy.org/es/index.html) para álgebra y análisis simbólico
    - [NumPy](https://numpy.org/) y [SciPy](https://scipy.org/) para cálculo numérico
    - [matplotlib](https://matplotlib.org/) para generar representaciones visuales de los resultados

## [Aula invertida](https://es.wikipedia.org/wiki/Aula_invertida)

flippedClassroomSchedule_castellano.png

- Teoría
    - Se lee, e intenta aplicar, **antes** del encuentro semanal.
    - Las dudas pueden aclararse **durante** el encuentro.
    - Se puede hacer consultas asincrónicas **fuera** del encuentro.
- Ejercicios
    - Se inician **antes** del encuentro en que se entregan.
    - Aclaradas consultas se terminan **durante** el encuentro en que se entregan.
    - Correcciones asincrónicas **fuera** del encuentro en que se entregan.

| Tiempo    | Modalidad tradicional | Aula invertida |
| -------- | ------- |------- |
| Sincrónico con los docentes  | Docente: expone nueva teoría.<br/> Alumno: poco tiempo para consultar. |  Alumno: termina ejercicios de la semana.<br/> Docente: atiende consultas que habiliten finalizar ejercicios. |
| Asincrónico<br/>(en casa o donde se pueda) | Alumno: repaso de teoría y resolver ejercicios. | Alumno: asimila nueva teoría e inicia resolución de ejercicios.

### Organizarse en un aula invertida

- Esta modalidad requiere **mucha regularidad** por parte del alumno
- Si se espera al encuentro sincrónico para:
  - leer teoría $\implies$ **no alcanza el tiempo**
  - empezar los ejercicios $\implies$ **no alcanza el tiempo**
  - ponerse _al día_ con lo adeudado de la semana anterior $\implies$ **no alcanza el tiempo**
- Para ayudar a manterse al día habrá una **seguimiento semanal**
  - **Nueva guía** de ejercicios cada semana
  - **Entrega obligatoria** de dos o más ejercicios vía Teams a las **21:50 hs** en cada encuentro sincrónico
- Por supuesto que puede fallarse en uno o más de los ejercicios que se entregan.
  - Es posible que haya faltado comprensión sobre algún aspecto de teoría $\implies$ consulta y/o repaso personal de teoría
  - Pero tal falla debe corregirse en una **segunda entrega**.

### Asistencia
- En encuentros sincrónicos se toma lista.
- Sí faltan por cuestiones médicas, un certificado permite descontarle.

### Entrega de ejercicios
- Cada unidad tiene su guía de ejercicios.
    - Deben entregarse la resoluciones de los que aparecerán en _Teams_ bajo _Tareas_.
- El día que se presenta la unidad podrán comenzar a mirarlos y resolverlos.
    - En el siguiente encuentro podrán realizar consultas adicionales
    - Los ejercicios se entregan antes de que comience el siguiente.
- Los encuentros sincrónicos son para consultas y finalizar ejercicios.
    - También pueden consultar en Teams durante la semana.

#### Esquema de entregas
> La guía de ejercicios N se publica en la semana N-1 $\rightarrow$ en N se entregan

- Pasado el día de entrega $\implies$ penalización sobre la nota
    - Por ello es importante que intenten avanzar en ellos, no es necesario que estén perfectos, una vez los envíen se los corregiremos y les pediremos las modificaciones necesarias para que puedan avanzar hasta que lo terminen satisfactoriamente.
    - Cuando termine la cursada entonces deberían haber logrado hacer **TODOS** los ejercicios asignados como _Tarea_.
